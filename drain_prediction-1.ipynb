```python
data = pd.read_csv(r"C:\Users\Edunet Foundation\Downloads\drain_data_100_bool.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
bool_columns = ['gas_value', 'rain_value', 'wf_value']
for col in bool_columns:
    data[col] = data[col].astype(int)
```

```python
data.head()
```

```python
features = ['gas_value', 'rain_value', 'temp_value', 'water_dist', 'wf_value']
X = data[features]
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
data['cluster_label'] = kmeans.fit_predict(X_scaled)
data.groupby('cluster_label').mean()
```

```python
data['drain_status'] = data['cluster_label'].map({0: 'blocked', 1: 'normal'})
```

```python
data.drop(columns=['cluster_label'], inplace=True)
```

```python
data.head()
```

```python
from sklearn.preprocessing  import LabelEncoder
le = LabelEncoder()
data['drain_status'] = le.fit_transform(data['drain_status'])   
```

```python
from sklearn.model_selection import train_test_split
X = data[features]
y = data['drain_status']    
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(x_train, y_train)   
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
accuracy = accuracy_score(y_test, y_pred)  
print("Accuracy:", accuracy) 
```

```python
print(classification_report(y_test, y_pred))
confusion_matrix(y_test, y_pred)
```

```python
import joblib
joblib.dump(scaler, "scaler.pkl")
joblib.dump(clf, 'drain_status_model.pkl')  
```

```python
model = joblib.load('drain_status_model.pkl')
def test_model(gas, rain, temp, dist, flow):
    # Prepare the input in the exact trained order
    # gas_value, rain_value, temp_value, water_dist, wf_value
    data = [[int(gas), int(rain), temp, dist, int(flow)]]
    columns = ['gas_value', 'rain_value', 'temp_value', 'water_dist', 'wf_value']
    test_df = pd.DataFrame(data, columns=columns)
    
    # Get the raw numerical prediction
    raw_pred = model.predict(test_df)[0]
    
    # 3. STRICT MAPPING: Only Blocked or Normal
    if raw_pred == 0:
        return "Blocked"
    else:
        return "Normal"

# --- RUN TESTS ---
print(f"Scenario 1 (No Flow + Gas): {test_model(False, False, 32.0, 500, False)}")
print(f"Scenario 2 (Flowing + Clear): {test_model(False, False, 27.0, 36, True)}")
```
